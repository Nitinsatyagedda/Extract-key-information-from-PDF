```python
import PyPDF2
import re
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
import nltk
nltk.download('punkt')
```

```python
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    return text
```

```python
def extract_key_information(text):

    key_info = {}

    # Extract sentences mentioning growth prospects, key changes, triggers, etc.
    sentences = sent_tokenize(text)

    # Regular expressions to identify key information
    growth_prospects_regex = re.compile(r'\b(?:growth prospects|future growth)\b', re.IGNORECASE)
    key_changes_regex = re.compile(r'\b(?:key changes|important changes)\b', re.IGNORECASE)
    triggers_regex = re.compile(r'\b(?:triggers|key triggers)\b', re.IGNORECASE)

    # Initialize variables to store key information
    growth_prospects = []
    key_changes = []
    triggers = []

    # Extract relevant sentences
    for sentence in sentences:
        if growth_prospects_regex.search(sentence):
            growth_prospects.append(sentence)
        if key_changes_regex.search(sentence):
            key_changes.append(sentence)
        if triggers_regex.search(sentence):
            triggers.append(sentence)

    key_info['growth_prospects'] = growth_prospects
    key_info['key_changes'] = key_changes
    key_info['triggers'] = triggers

    return key_info
```

```python
if __name__ == "__main__":
    pdf_path = "SJS Transcript Call.pdf"
    text = extract_text_from_pdf(pdf_path)
    key_info = extract_key_information(text)
    print("Growth Prospects:")
    for sentence in key_info['growth_prospects']:
        print("-", sentence)
    print("\nKey Changes:")
    for sentence in key_info['key_changes']:
        print("-", sentence)
    print("\nTriggers:")
    for sentence in key_info['triggers']:
        print("-", sentence)
```
